```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['1-Day MYW Seasonal Base', '1-Day MYW Seasonal Park Hopper', '1-Day MYW Seasonal with Park Hopper Plus Option']
adult_prices = [140.75, 217.00, 238.75]
child_prices = [135.25, 211.50, 233.50]

# Set the width of the bars
bar_width = 0.3

# Set the position of the bars on the y-axis
r = np.arange(len(categories))

# Create horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(r - bar_width / 2, adult_prices, color='skyblue', height=bar_width, edgecolor='grey', label='Adult (10+)')
plt.barh(r + bar_width / 2, child_prices, color='lightcoral', height=bar_width, edgecolor='grey', label='Child (3-9)')

# Adding labels and title
plt.title('Magic Your Way Seasonal Ticket Prices', fontsize=16)
plt.xlabel('Price ($)', fontsize=14)
plt.ylabel('Ticket Type', fontsize=14)
plt.yticks(r, categories)
plt.legend()

# Adding exact prices within the chart borders
for i in range(len(categories)):
    plt.text(adult_prices[i] - 20, r[i] - bar_width / 2, f"${adult_prices[i]}", va='center', ha='right', color='black')
    plt.text(child_prices[i] - 20, r[i] + bar_width / 2, f"${child_prices[i]}", va='center', ha='right', color='black')

# Show plot
plt.tight_layout()
plt.show()
```
